```python
import pandas as pd

df_consumo = pd.read_csv("src/data/Coffee_domestic_consumption.csv")
```

```python
year_columns = [col for col in df_consumo.columns if '/' in str(col)]
print(year_columns[:10])  # ver las primeras columnas de años
```

```python
consumo_long = pd.melt(
    df_consumo,
    id_vars=['Country', 'Coffee type'],   # columnas que identifican cada fila
    value_vars=year_columns,              # columnas de años
    var_name='Year_Period',               # nombre de la columna temporal
    value_name='Domestic_Consumption'     # nombre de la variable
)
```

```python
consumo_long['Year'] = consumo_long['Year_Period'].str.split('/').str[0].astype(int)
consumo_long = consumo_long.drop(columns=['Year_Period'])
```

```python
consumo_long = consumo_long.rename(columns={'Country': 'País'})
consumo_long['Domestic_Consumption'] = pd.to_numeric(consumo_long['Domestic_Consumption'], errors="coerce")
```

```python
consumo_long.to_csv("src/data/Coffee_domestic_consumption_long.csv", index=False)
consumo_long
```
