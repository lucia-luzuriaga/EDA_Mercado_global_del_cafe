```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
consumo = pd.read_csv("./src/data/Coffee_domestic_consumption.csv")
export = pd.read_csv("./src/data/Coffee_export.csv")
import_ = pd.read_csv("./src/data/Coffee_import.csv")
re_export = pd.read_csv("./src/data/Coffee_re_export.csv")
importers = pd.read_csv("./src/data/Coffee_importers_consumption.csv")
inventory = pd.read_csv("./src/data/Coffee_green_coffee_inventorie.csv")
```

‚ÄúSe va a crear una funci√≥n auxiliar para realizar una exploraci√≥n de todos los datasets, permitiendo identificar r√°pidamente su estructura, calidad y posibles problemas antes del an√°lisis.‚Äù

```python
def explore_df(df, name):

    print(f"DATASET: {name}")
  
    # Primeras filas
    print("\nüîπ Primeras filas:")
    display(df.head())
    
    # Tama√±o
    print(f"\nüîπ Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
    
    # Info
    print("\nüîπ Informaci√≥n general:")
    df.info()
    
    # Valores nulos
    print("\nüîπ Valores nulos por columna:")
    display(df.isna().sum())
    
    # Duplicados
    print(f"\nüîπ Filas duplicadas: {df.duplicated().sum()}")
    
    # Estad√≠sticas descriptivas
    print("\nüîπ Estad√≠sticas descriptivas:")
    display(df.describe(include="all"))
```

```python
explore_df(consumo, "Consumo")
explore_df(export, "Export")
explore_df(import_, "Import")
explore_df(re_export, "Re-Export")
explore_df(importers, "Consumo de Importadores")
explore_df(inventory, "Green Coffee Inventory")
```
